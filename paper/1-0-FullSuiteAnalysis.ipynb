# Analysis on Suite A


```python
incs = [0, 15, 30, 45, 60, 75, 90]

tempdir = 'temp/fullsuite/A/'


for suite_index in range(1, 7):
    orbit_set = []

    for inc in incs:
        orbit_set.append(grp.OrbitContainer.load(f'{tempdir}A{suite_index}_{inc}.grp'))

    plt.figure(figsize=(8, 6))
    for i, container in enumerate(orbit_set):
        times, mdisk = grp.Mdisk(container)
        plt.plot(times, mdisk, label=f'inc = {incs[i]}')
    plt.legend()
    plt.title(f'A{suite_index}')
    plt.tight_layout()
    plt.savefig(f'figures/fullsuite/A/A{suite_index}_mdisk_evol.png')
```
